```python
import pandas as pd
df = pd.read_csv("Recorridos.csv")
df.drop(df.columns[[0, 3,4,5,6,7,8]], axis = 1, inplace = True) 
df.head()
```

```python
df.rename(columns={'bici_Fecha_hora_retiro':'Fecha',
                          'bici_tiempo_uso':'uso'},
                 inplace=True)
```

```python
import datetime
def daynumber(date_time_str):
    date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S')
    return date_time_obj.weekday()

df['Fecha'] = df['Fecha'].apply(daynumber)
```

```python
df
```

```python
df["uso"] = df['uso'].astype('str')
from pytimeparse.timeparse import timeparse
df["uso"] = df["uso"].apply(timeparse)
```

```python
df
```

```python
df = df.groupby(['Fecha']).sum()
```

```python
df["uso"] = df["uso"].apply(lambda x: (x / 60)/60)
```

```python
df
```

```python
df.index.name = 'weekday'
df.rename(index={0: 'Monday'}, inplace = True)
df.rename(index={1: 'Tuesday'}, inplace = True)
df.rename(index={2: 'Wednesday'}, inplace = True)
df.rename(index={3: 'Thursday'}, inplace = True)
df.rename(index={4: 'Friday'}, inplace = True)
df.rename(index={5: 'Saturday'}, inplace = True)
df.rename(index={6: 'Sunday'}, inplace = True)
df.rename(columns={'uso':'use (hours)'},
                 inplace=True)
df = df.reset_index()
df
```

```python
import matplotlib.pyplot as plt
import matplotlib as mp
import seaborn as sns
import numpy as np

%matplotlib inline

plt.style.available
```

```python

import matplotlib.pyplot as plt
import matplotlib as mp
import numpy as np
 
fig, ax = plt.subplots(figsize=(10,6))

# Colorize the graph based on likeability:
likeability_scores = np.array(df["use (hours)"])
 
data_normalizer = mp.colors.Normalize()
color_map = mp.colors.LinearSegmentedColormap(
    "my_map",
    {
        "red": [(0, 1.0, 1.0),
                (1.0, .5, .5)],
        "green": [(0, 0.5, 0.5),
                  (1.0, 0, 0)],
        "blue": [(0, 0.50, 0.5),
                 (1.0, 0, 0)]
    }
)
 
# Map xs to numbers:
N = len(df["use (hours)"])

 
# Plot a bar graph:
plt.bar(
    df["weekday"],
    df["use (hours)"],
    align="center",
    color=color_map(data_normalizer(likeability_scores))
)
 
plt.xticks(rotation=90)

plt.xticks(df["weekday"])
plt.title('Total hours of bicycle use per weekday during 2018 \n', fontsize=20, fontweight='black', color = '#333F4B')
ax.set_ylabel('Hours of use', fontsize=18, fontweight='black', color = '#333F4B')
ax.set_xlabel('Weekday', fontsize=18, fontweight='black', color = '#333F4B')

plt.show()
```
